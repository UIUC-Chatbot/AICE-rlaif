```python
from exa_py import Exa

search_response_dict = {}
title = []
url = []
score = []
text = []

# instantiate the Exa client
# gmail key
# exa = Exa("b4172bbd-776b-4248-83c9-230972418f66")

# uiuc key
exa = Exa("015c8a7f-0179-47e5-999b-6815df432b8a")

for i in range(0,100):
  search_response = exa.search_and_contents(dataset['train'][i]['conversations'][0]['value'], use_autoprompt=True)

for r in search_response.results:
    title.append(r.title)
    url.append(r.url)
    score.append(r.score)
    text.append(r.text)
```

```python
search_response_dict = {'Title': title, "url": url, "Score": score, "Text" : text}
print(search_response_dict['Title'])
print(len(search_response_dict['Title']))
```

```python
import json
with open('search_response.json', 'w') as f:
    # Write the dictionary to the file as JSON
    json.dump(search_response_dict, f)
```
